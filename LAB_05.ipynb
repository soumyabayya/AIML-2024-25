```python
df.drop(['alive','alone','deck'],axis=1,inplace=True)
df.dropna(inplace=True)
df.head()
```

```python
sex=pd.get_dummies(df['sex'],drop_first=True)
embark_town=pd.get_dummies(df['embark_town'],drop_first=True)
```

```python
df.drop(['sex','embark_town'],axis=1,inplace=True)
df=pd.concat([df,sex,embark_town],axis=1)
df.head()
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(df.drop('survived',axis=1),df['survived'],test_size=0.38,random_state=111)
```
